# Djangodaki static dosyaları kullanma

https://docs.djangoproject.com/en/5.0/howto/static-files/


linkteki dökümantasyonu biraz okuyalım ve gerekliliklerine bakalım.

şimdi static klasörü içerisine style.css dosyamızı oluşturalım ve p elementlerinin arkaplan rengini deneme amaçlı kırmızı yapalım.

```python
p {
    background: red;
}
```

bunu yaptığımız zaman django direkt olarak gidip static dosyalarımızın altına bakacak. style.css i projemize dahil etmek için aşağıdaki gibi bir tanımlama yapmamız gerekiyor.

```python
<link rel="stylesheet" href="../static/style.css"> 
```

Bu şekilde style.css'i kullanmış olduk. sayfa kaynağından da görüntüleyebiliriz. Ama biz django projelerimizde bu yapıyı kullanmak yerine djangonun bize static dosyalar için sunmuş olduğu yöntemi kullanacağız. Bu yöntem static dosyalarımızı otomatik olarak daha dinamik bir şekilde bulmamızı sağlıyor. Aşağıdaki load static işlemi şu an için sadece app'lerimiz içindeki static klasörünü arıyor

```python
{% load static %}
```

```python
<link rel="stylesheet" href="{% static 'style.css' %}"> 
```

Biz yine style.css dosyamızı css diye bir klasör içerisine taşıyabiliriz. Bunu yaparsak da aşağıdaki kod düzeltmesini yapmamız gerekir.

```python
<link rel="stylesheet" href="{% static 'css/style.css' %}"> 
```

Biz bu static klasörünü app'lerin içerisine değil de direkt anadizine oluşturmak istersek ne yapacağız? önce static klasörünü anadizinde oluşturabiliriz. ve tanımlaması için Dökümana bakabiliriz.

```python
STATICFILES_DIRS = [
    BASE_DIR / "static",
]
```
